```
class maquinaDeTuring:
  def __init__(self, alfabeto, n_estados, estado_inicial, estados_aceptacion):
    self.alfabeto = alfabeto
    self.alfabeto.append(' ')
    self.n_estados = n_estados
    self.estado_inicial = estado_inicial
    self.estados_aceptacion = estados_aceptacion
    self.transiciones = {}
  
  def agregarTransicion(self, estado_inicial, simbolo_inicial, estado_final, simbolo_final, desplazamiento):
    assert simbolo_final in self.alfabeto and simbolo_final in self.alfabeto
    self.transiciones[(estado_inicial, simbolo_inicial)] = (estado_final, simbolo_final, desplazamiento)
  
  def procesarEntrada(self, cadena):
    estado = self.estado_inicial
    indice = 0
    while True:
      assert cadena[indice] in self.alfabeto
      if (estado, cadena[indice]) in self.transiciones:
        ef, sf, d = self.transiciones[(estado, cadena[indice])]
        cadena[indice] = sf
        estado = ef
        indice += d
        if estado in self.estados_aceptacion:
          return True
      else:
        return False
```

```
def entero_a_lista_bin(n):
  return list(bin(n)[2:]+' ')
```

```
class mt_par(maquinaDeTuring):
  def __init__(self):
    super().__init__(['0', '1'], 3, 0, [2])
    self.agregarTransicion(0,'1',0,'1',1)
    self.agregarTransicion(0,'0',0,'0',1)
    self.agregarTransicion(0,' ',1,' ',-1)
    self.agregarTransicion(1,'0',2,'0',0)

  def esPar(self, numero):
    return self.procesarEntrada(entero_a_lista_bin(numero))    
```
